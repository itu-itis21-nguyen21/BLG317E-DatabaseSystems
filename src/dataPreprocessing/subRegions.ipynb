```python
import pandas as pd
```

```python
data = pd.read_csv("../database/original/countries.csv", delimiter=',')
data
```

```python
data = data.drop(columns=["name", "alpha-2", "alpha-3", "country-code", "iso_3166-2", "intermediate-region", "intermediate-region-code", "region"])
data
```

```python
data.dtypes
```

```python
data["region-code"] = data["region-code"].astype("Int64")
data["sub-region-code"] = data["sub-region-code"].astype("Int64")
data
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data.dropna(inplace=True)
data
```

```python
data.to_csv("../database/fixed/subRegions.csv", index=False)
```
